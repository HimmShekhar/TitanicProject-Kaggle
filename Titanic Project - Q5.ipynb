#5.) What factors helped someone survive the sinking?

```python
import pandas as pd
from pandas import Series, DataFrame

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic_df = pd.read_csv('train.csv')
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0:'No',1:'Yes'})
```

```python
titanic_df.head()
```

```python
# Plotting Survivor count

sns.catplot(x='Survivor',kind='count',data=titanic_df)
```

```python
#value_counts() : count the unique values for passed object
titanic_df['Survivor'].value_counts()
```

```python
# Defining a method to check if the passenger was a child.

def male_female_child(passenger):
    age,sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
#Apply Function : One alternative to using a loop to iterate over a DataFrame is to use the pandas .apply() method. This function acts as a map() function in Python. It takes a function as an input and applies this function to an entire DataFrame.

titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
# # Plotting Survivor count on basis of Person(Child, Male and Female) and Class they were travelling

sns.catplot(x='Pclass',y='Survived',hue='person',data=titanic_df,kind='point')
```

```python
#Seaborn's lmplot is a 2D scatterplot with an optional overlaid regression line. This is useful for comparing numeric variables.
sns.lmplot(x='Age',y='Survived',data=titanic_df)
```

```python
#Plotting to analyze survival on the basis of age and class of passengers. 
sns.lmplot(x='Age',y='Survived',hue='Pclass',data=titanic_df)
```

```python
gen = [10,20,30,40,50,60,70,80]

sns.lmplot(x='Age',y='Survived',hue='Pclass',data=titanic_df,palette='winter',x_bins=gen)
```

```python
#Plotting to analyze survival on the basis of age and gender of passengers. 
sns.lmplot(x='Age',y='Survived',hue='Sex',data=titanic_df,palette='winter',x_bins=gen)
```

Conclusion:

1. 342 passengers survived while rest 549 were unfortunate.
2. More female and children survived than male.
3. Passengers travelling in first class were more in number of survivors followed by second and third class respectively.
4. The older people were less in number of survivors.
